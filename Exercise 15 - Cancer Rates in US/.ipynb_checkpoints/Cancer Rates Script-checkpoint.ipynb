## Sub Challenge 1: Assessing Conditions for Linear Regression with POP, PCI, and MHI

To practice what we learned in the first half of Chapter 7, lets start by assessing the conditions for linear regression. Let’s first visualise the data:

```r
par(mfrow=c(1,2))

plot(PCI~POP, 
     data=cancer,
     main = "Per Capita Income by Pop",
     xlab = "Population",
     ylab = "Per Capita Income")

plot(MHI~POP, 
     data=cancer,
     main = "Median House Income by Pop",
     xlab = "Population",
     ylab = "Median Household Income")
```

Let's transform the values so that it becomes easier to spot trends.

```r
cancer$ln_POP <-log(cancer$POP)
cancer$ln_PCI <-log(cancer$PCI)
cancer$ln_MHI <-log(cancer$MHI)
```

```r
par(mfrow=c(1,2))

plot(ln_PCI~ln_POP, 
     data=cancer,
     main = "Per Capita Income by Pop",
     xlab = "Population (ln)",
     ylab = "Per Capita Inc. (ln)")

plot(ln_MHI~ln_POP, 
     data=cancer,
     main = "MHI by Pop",
     xlab = "Population (ln)",
     ylab = "Med. Houshold Inc. (ln)")
```

That seems like an improvement in that the scatterplots seem a bit more straight and the outliers in thescatterplots prior to transforming the data seem to have disappeared. At the same time, the clouds of points seems to be rather wide and do not closely follow any line we might draw through the cloud. Let’s proceed toestimate linear models using the transformed variables.

## Sub Challenge: Explore Linear Regression and R-Squared

We can now estimate linear models for the two pairs of variables and then store the results in two new object we will call lm.ln_PCI and lm.ln_MHI. We can then print the contents and see the estimated y-intercept as well as the slope of the line of best fit for the data.

```r
lm.ln_PCI <-lm(ln_PCI~ln_POP, data=cancer)
lm.ln_PCI
```

```r
lm.ln_MHI <-lm(ln_MHI~ln_POP, data=cancer)
lm.ln_MHI
```

```r
summary(lm.ln_PCI)$r.squared
```

```r
summary(lm.ln_MHI)$r.squared
```

So it appears that the model lm.ln_PCI accounts for about 4.4% of variation in ln_PCI while the model lm.ln_MHI accounts for about 6.9% of the variation in ln_MHI. That is to say that 95.6% of the variation remains unaccounted for by the first model while 93.1% remains unaccounted for by the second model.

```r
r.ln_PCI <-cor(cancer$ln_POP, cancer$ln_PCI, use="complete.obs")
r.ln_PCI^2
```

```r
r.ln_MHI <-cor(cancer$ln_POP, cancer$ln_MHI, use="complete.obs")
r.ln_MHI^2
```

## Sub Challenge 3: Examining the Residuals

Whereas taking the natural logarithm of POP, PCI, and MHI seemed to address initial concerns about the straight enough and outlier conditions when we looked at the scatter plots, it is important to examine themodel residuals in order to confirm that the straight enough and outlier conditions are met.

Perhaps the simplest way to do so is to. . .make a picture. As you will recall from the R-tutorial as part of today’s prep-material, we can do this using the plot() command once we have estimated a linear model. Let’s now plot and check the residuals from our two models:

```r
par(mfrow=c(1,2))

plot(lm.ln_PCI,
     which=1,
     main = "(ln) PCI by POP",
     sub.caption = "(ln) Per Capita Income")

plot(lm.ln_MHI, 
     which=1,
     main = "(ln) MHI by POP",
     sub.caption = "(ln) Median Household Income")
```

## Challenge 2

Since the counties vary so widely in population, it’s better for us to look at the rates rather than the number of cases.

### Step 4: Add a column called RATE that gives the number of cases per 100,000 residents.
Look at the distribution of the variable RATE.

```r
# CALCULATE THE RATE PER 100,000 PEOPLE AND FIT THAT INTO THE NEW COLUMN

cancer$RATE <- (cancer$CASES / cancer$POP) * 100000

head(cancer)
hist(cancer$RATE,
    main = "Histogram of Lung Cancer Rate",
    xlab = "Lung Cancer Cases per 100,000 Population",
    col = "pink")
```

### Step 5: Make two histograms: one of counties where the rate is greater than 200 and one where the rate is at most 200.

```r
# CREATE TWO HISTOGRAMS
par(mfrow = c(1,2))

hist(cancer$RATE[cancer$RATE > 200],
    freq = FALSE,
    col = "pink",
    main = "Histogram of Cancer Rate (>200)",
    xlab = "Cancer Cases per 100,000 Pop")

hist(cancer$RATE[cancer$RATE > 200],
    freq = FALSE,
    col = "pink",
    main = "Histogram of Cancer Rate (>200)",
    xlab = "Cancer Cases per 100,000 Pop")
```

## Challenge 3

Let’s further focus on counties where the rate is extremely high and extremely low.

### Step 6: Create a new data frame consisting of counties where the rate is at least 500.
- How many are there?
- Looking at this data frame, is there anything that distinguishes these counties?
- What about for counties with very low rates (i.e. less than 20)?

```r
# EXAMINE THE DATAFRAMES

mean(cancer$POP, na.rm = TRUE)
mean(high$POP, na.rm = TRUE)
mean(low$POP, na.rm = TRUE)
```

Interesting: it appears that high rate counties have much smaller populations on average (about 7,000) than all counties or low rate counties (both have about or above 100,000). What could explain the low cancer rates in large counties (cities)?

```r
# COMPARE THE MEAN OVER65 OF ALL COUNTIES WITH THOSE OF HIGH AND LOW

mean(cancer$OVER65 / cancer$POP, na.rm = TRUE)
mean(high$OVER65 / high$POP, na.rm = TRUE)
mean(low$OVER65 / low$POP, na.rm = TRUE)
```

Interesting: the low rate counties also have low proportion of persons over 65.

## Challenge 4
We noticed that the counties with high lung cancer rates generally had low populations, while counties with more average lung cancer rates tend to have larger populations. This is a manifestation of regression to the mean.

But the relationship isn’t as simple for the counties with low rates. It seems that at least some of the counties with low rates are very populous but also have a younger population. Let’s investigate further.

### Step 7: How does the average per capita income of low-rate counties compare to the average of all counties?

### Step 8: How does the average median household income of low-rate counties compare to the average of all counties?

```r
mean(low$MHI, na.rm = TRUE)

mean(cancer$MHI, na.rm = TRUE)
```

### Step 9: Make plots and linear models for RATE as a function of PCI and of MHI.
- If there are outliers, then how should they be dealt with?

```r
par(mfrow = c(1,2))

plot(RATE ~ PCI, 
    data = cancer, 
    main = 'Cancer Rate by Per Capita Income',
    xlab = 'Per Capita Income',
    ylab = 'Cancer Cases per 100k')

plot(RATE ~ MHI, 
    data = cancer, 
    main = 'Cancer Rate by Median Household Income',
    xlab = 'Median Household Income',
    ylab = 'Cancer Cases per 100k')

par(mfrow = c(1,1))
```

```r
par(mfrow=c(2,1))

plot(RATE ~ PCI, 
     data=cancer[cancer$RATE <= 200,], 
     main = "Cancer Rate (<=200) by Per Capita Income",
     xlab = "Per Capita Income", 
     ylab = "Cancer Cases per 100k")

abline(lm(RATE ~ PCI, data=cancer[cancer$RATE <= 200,]), col="red")

plot(RATE ~ MHI, 
     data=cancer[cancer$RATE <= 200,], 
     main = "Cancer Rate (<=200) by Median Household Income",
     xlab = "Median Household Income", 
     ylab = "Cancer Cases per 100k")

abline(lm(RATE ~ MHI, data=cancer[cancer$RATE <= 200,]), col="red")

par(mfrow=c(1,1))
```

## Conclusion

Taking a step back, what have we learned?

- Small samples have higher variability than larger samples.
- All else equal, we would expect the small samples (counties with low populations) to be over-represented at the high end and the low end.
- Hidden variable in this data set: young residents in high population counties.
- Over the long run, though, individual outliers will revert toward the mean.
