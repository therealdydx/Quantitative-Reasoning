### Step 1: Check for NA values in the columns

Do we have NAs in the columns weight and height? If yes, how much? Remove the values.

```r
if (any(is.na(nhanes$height))) {
    print(TRUE)
}

if (any(is.na(nhanes$weight))) {
    print(TRUE)
}
```

```r
# CHECK THE AMOUNT OF NA VALUES IN THE DATASET

length(nhanes$height[is.na(nhanes$height)])
length(nhanes$weight[is.na(nhanes$weight)])

# NOW REMOVE THE NA VALUES IN THE DATASET

nhanes <- nhanes[!(is.na(nhanes$weight) | is.na(nhanes$height)), ]
```

### Step 2: Create visualizations of distributions of variables weight and height

Weâ€™ll look at people from 20 years to 60 years old. Reason is that the definition of obesity is different for younger people and because causes and consequences of weight issues start to change when people get older.

- Create a new data frame with the subset of people >20y and <60 year old

- Visualize the distribution of the variables weight and heigth

```r
# CREATE A NEW DATAFRAME

nhanes <- nhanes[nhanes$age_yrs > 20 & nhanes$age_yrs < 60, ]
```

```r
# Visualize the distrbutions of the weight and height

par(mfrow = c(1,2))

hist(nhanes$height, 
    col = "purple",
    freq = FALSE,
    xlab = "height (cm)")

hist(nhanes$weight,
    col = "red",
    freq = FALSE,
    xlab = "weight (kg)")
```

The distribution of the variable weight appears to be right skewed while that of height seems fairly normal. Transform the weight and plot the data in a histogram to see if it gets us a more normal distribution.

### Step 3: Transform the data and re-visualize the plot to see if there is a distribution

```r
# TRANSFORM THE DATASET WITH LOGARITHMIC TO SEE IF THERE IS A NORMAL DISTRIBUTION

nhanes$log_weight <- log(nhanes$weight)
nhanes$log_height <- log(nhanes$height)

par(mfrow = c(2,1))

hist(nhanes$log_height,
     col = 'purple',
     freq = FALSE,
     xlab = 'height (cm)')

hist(nhanes$log_weight,
     col = 'red',
     freq = FALSE,
     xlab = 'weight (kg)')
```

### Step 4: Find the relationship between weight and height

- Use the function lm() to fit a regression line
- Plot out a scatterplot
- Add the regression line


```r
lm_hw <- lm(log_weight ~ height, data = nhanes)
lm_hw
```

Not very surprisingly, there seems to be a positive association between the height of a person and that personâ€™s weight. Specifically, with each centimeter increase in height, the average person gains a bit over log(weight) = 0.012, which is e^0.012 = exp(0.012) = 1.01 kg.

```r
plot(log_weight ~ height,
     data = nhanes,
     col = "red",
     xlab = "height (cm)",
     ylab = "log(weight) (kg)")

abline(lm_hw, col = "blue")
```

### Step 5: Comparing height-weight relationships among groups

The survey data includes information on gender, ethnic groups, and income groups. If the height-weight relationship differs between the categories within these groups, that would potentially explain part of the variation we observe in the scatter plot. Letâ€™s start by investigating if there are gender differences in the height-weight relationship.

- Create two different data frames, one with men and the other with women, using subsetting.

- Check the distribution of weight and of log(weight) of men and women using histograms. Plot the 4 histograms in one display.

- Fit height-weight linear relationships on both relationships and inspect the output

- Plot the data points and regression lines of both men and women in one single graph

```r
men <- nhanes[nhanes$gender == 'M', ]

women <- nhanes[nhanes$gender == "F", ]
```

```r
par(mfrow = c(2,2))

hist(men$log_weight,
    col = "cyan",
    freq = FALSE,
    xlab = "log of the weight",
    main = "men - log of weight")

hist (women$log_weight,
    col = "lightblue",
    freq = FALSE,
    xlab = "log of the weight",
    main = "women - log of weight")

hist(men$weight,
    col = "darkgreen",
    freq = FALSE,
    xlab = "weight",
    main = "men - weight")

hist(women$weight,
    col = "green",
    freq = FALSE,
    xlab = "weight",
    main = "women - weight")
```

### Step 6: Compare height-weight relationship among groups

```r
lm_women <- lm(log_weight ~ height, data = women)

lm_men <- lm(log_weight ~ height, data = men)

lm_women; lm_men
```

```r
plot(log_weight ~ height, 
    data = men,
    col = "lightblue",
    xlab = "height(cm)",
    ylab = "log of weight(kg)")

points(log_weight ~ height,
    data = women,
    col = "darkgreen",
    xlab = "height(cm)",
    ylab = "log of weight(kg)")

abline(lm_men, col = "darkblue", lwd = 3)
abline(lm_women, col = "darkgreen", lwd = 3)
```

There seems to be some gender differences in the regression lines, but very small. There are clear gender differences discernible in this graph though, for example men tend to be taller. Less evident is the larger variation in the weight:height ratio among women.

```r
men_00 <- men[men$year == "2000", ]

men_18 <- men[men$year == "2018", ]

women_00 <- women[women$year == "2000", ]

women_18 <- women[women$year == "2018", ]
```

```r
lm_men_00 <- lm(log_weight ~ height, data = men_00)
lm_men_00

lm_women_00 <- lm(log_weight ~ height, data = women_00)
lm_women_00

lm_men_18 <- lm(log_weight ~ height, data = men_18)
lm_men_18

lm_women_18 <- lm(log_weight ~ height, data = women_18)
lm_women_18
```

So, what did we find? There is an obvious relationship between the height and weight of a person. There is, however, a large variation among people. There are many people in this data set with a height of 1.50 m that have a similar weight as someone of 1.85 m tall, either because the former has a very high body mass index or because the latter has a very low body mass index.

There are many reason for this variation, one of which is stochasticity, differences in diet that in turn may be related to a multitude of variables such as income-related access to healthier food, education, cultural background, genetic variation, etc. So investigating a linear relationship between two variables is often just the first step in understanding what drives that relationship and what drives variation in that relationship.

Our analyses suggest some gender and year related variation in the height-weight relationship, but the graph makes clear that those differences are very small compared with the total amount of variation in the data.

### Step 9: Other drivers of variation of height weight relationship

The survey data includes information on gender, ethnic groups, and income groups. If the height-weight relationship differs between the categories within these groups, that would potentially explain part of the variation we observe in the scatter plot. We looked at gender but there are more possible drivers of variation in this data set. Letâ€™s start by investigating if there are differences in the height-weight relationship of the major ethnic groups.

- What ethnic groups are there?
- Create different data frames for each group, using subsetting.
- Fit regression lines per group and take a look at the regression slope and intercept.
- Plot the data points and regression lines of all groups in one single graph.

```r
table(nhanes$ethn, nhanes$gender)
```

```r
f_nhw <- nhanes[nhanes$ethn == "Non-Hispanic White" & nhanes$gender == "F", ]
f_nhb <- nhanes[nhanes$ethn == "Non-Hispanic Black" & nhanes$gender == "F", ]
f_ma <- nhanes[nhanes$ethn == "Mexican American" & nhanes$gender == "F", ]
f_oh <- nhanes[nhanes$ethn == "Other Hispanic" & nhanes$gender == "F", ]
f_mix <- nhanes[nhanes$ethn == "Other Race and multi" & nhanes$gender == "F", ]

m_nhw <- nhanes[nhanes$ethn == "Non-Hispanic White" & nhanes$gender == "M", ]
m_nhb <- nhanes[nhanes$ethn == "Non-Hispanic Black" & nhanes$gender == "M", ]
m_ma <- nhanes[nhanes$ethn == "Mexican American" & nhanes$gender == "M", ]
m_oh <- nhanes[nhanes$ethn == "Other Hispanic" & nhanes$gender == "M", ]
m_mix <- nhanes[nhanes$ethn == "Other Race and multi" & nhanes$gender == "M", ]
```

```r
# CREATE LINEAR MODELS FOR ALL OF THE RELATIONSHIPS

lm_f_nhw <- lm(log_weight ~ height, data = f_nhw)
lm_f_nhb <- lm(log_weight ~ height, data = f_nhb)
lm_f_ma <- lm(log_weight ~ height, data = f_ma)
lm_f_oh <- lm(log_weight ~ height, data = f_oh)
lm_f_mix <- lm(log_weight ~ height, data = f_mix)

lm_m_nhw <- lm(log_weight ~ height, data = m_nhw)
lm_m_nhb <- lm(log_weight ~ height, data = m_nhb)
lm_m_ma <- lm(log_weight ~ height, data = m_ma)
lm_m_oh <- lm(log_weight ~ height, data = m_oh)
lm_m_mix <- lm(log_weight ~ height, data = m_mix)


m_ethn <- list(m_nhw, m_nhb, m_ma, m_oh, m_mix)
f_ethn <- list(f_nhw, f_nhb, f_ma, f_oh, f_mix)

m_lm_list <- list(lm_m_nhw, lm_m_nhb, lm_m_ma, lm_m_oh, lm_m_mix)
f_lm_list <- list(lm_f_nhw, lm_f_nhb, lm_f_ma, lm_f_oh, lm_f_mix)

colors <- c("blue", "darkgreen", "orange", "purple", "red")
```

```r
plot(log_weight ~ height,
    data = m_nhw,
    col = "blue",
    xlab = "height",
    ylab = "log of weight")

par(mfrow = c(1,1))

for (i in 2:5) {
    df = m_ethn[[i]]
    points(df$log_weight ~ df$height,
           col = colors[i])
}

for (i in 1:5) {
    abline(m_lm_list[[i]], col = colors[i], lwd = 3)
}
```

```r
plot(log_weight ~ height,
    data = f_nhw,
    col = "blue",
    xlab = "height",
    ylab = "log of weight")

for (i in 2:5) {
    df = f_ethn[[i]]
    points(df$log_weight ~ df$height,
           col = colors[i])
}

for (i in 1:5) {
    abline(f_lm_list[[i]], col = colors[i], lwd = 3)
}
```
