### Introduction

Flipping 10 times a coin that was thought to be fair results in 7 heads and 3 tails.

- Show that this event is not statistically significant.
- Find out how far you have to reduce the total number of tosses for 7 heads to be statistically significant.

### Step 1: Show that this event is not statistically significant

```r
# CREATE NUMBER OF TRIALS
trials <- 10000

# INITIALIZE AN ARRAY
responses <- numeric(trials)

# NUMBER OF TOSSES
tosses <- 10

# APPEND VALUES INTO ARRAY
for (i in 1:trials) {
    
    # SIMULATE THE RESULT FOR 10 TOSSES
    results <- sample(c("H", "T"), size = tosses, replace = TRUE)
    
    # APPEND VALUE INTO THE ARRAY
    responses[i] <- sum(results == "H")
}
```

```r
hist(responses, 
    freq = FALSE,
    breaks = seq(-0.5, tosses + 0.5, 1),
    col = "lightblue")

abline(v = quantile(responses, prob = c(0.025, 0.975)),
      col = "orange",
      lwd = 2)

abline(v = 7, 
       col = "blue",
      lwd = 2)
```

### Step 2: Find out how far you have to reduce the total number of tosses for 7 heads to be statistically significant

```r
# CREATE NUMBER OF TRIALS
trials <- 10000

# INITIALIZE AN ARRAY
responses <- numeric(trials)

# CREATE 4 PLOTS
par(mfrow = c(2,2))

tosses <- 7

# CREATE NESTED FOR LOOPS
for (j in 7:10) {
    
    # APPEND VALUES INTO ARRAY
    for (i in 1:trials) {
    
        # SIMULATE THE RESULT FOR 10 TOSSES
        results <- sample(c("H", "T"), size = tosses, replace = TRUE)
    
        # APPEND VALUE INTO THE ARRAY
        responses[i] <- sum(results == "H")
    }
    
    # PLOT A HISTOGRAM
    hist(responses, 
        col = "lightblue",
        main = paste("Total of", tosses, "tosses"),
        breaks = c(seq(-0.5, tosses + 0.5, 1)),
        freq = FALSE)
    
    # PLOT STRAIGHT LINES FOR FINDING TAILS
    abline(v = quantile(responses, prob = c(0.025, 0.975)),
          col = "orange",
          lwd = 2)
    
    # PLOT THE 7 HEADS RESULT LINE
    abline(v = 7, 
          col = "blue",
          lwd = 2)
    
    tosses <- tosses + 1
}
```

We have two coins that are both believed to be fair. Flipping the two coins simultaneously 10 times results in them both landing heads once; the other 9 times they were both tails, or one of each.

- Show that this event is not statistically significant.
- How far do you need to increase the total number of coin flips for only one double head to be statistically significant?

### Step 3: Show that this event is not statistically significant

```r
# NUMBER OF TRIALS
trials <- 10000

# INITIALIZE ARRAY
responses <- numeric(trials)

# DEFINE NUMBER OF TOSSES
tosses <- 10

# CREATE A FOR LOOP
for (i in 1:trials) {
    
    # CREATE TOSS 1
    results <- sample(c("HH", "HT", "TH", "TT"), size = tosses, replace = TRUE)
    
    # APPEND VALUES INTO ARRAY
    responses[i] <- sum(results == "HH")
}
```

```r
# PLOT A HISTOGRAM
hist(responses,
    breaks = c(seq(-0.5, 10.5, 1)),
    col = "lightblue")

# PLOT VERTICAL LINES
abline(v = quantile(responses, prob = c(0.025, 0.975)),
      col = "orange",
      lwd = 2)

abline(v = 1,
      col = "blue",
      lwd = 2)
```

### Step 4: How far do you need to increase the total number of coin flips for only one double head to be statistically significant?

```r
# NUMBER OF TRIALS
trials <- 10000

# INITIALIZE ARRAY
responses <- numeric(trials)

# CREATE 2 PLOTS
par(mfrow = c(2,1))

# CREATE A FOR LOOP
for (tosses in 19:20) {
    
    for (i in 1:trials) {
            
        # CREATE TOSS 1
        results <- sample(c("HH", "HT", "TH", "TT"), size = tosses, replace = TRUE)
    
        # APPEND VALUES INTO ARRAY
        responses[i] <- sum(results == "HH")
    }
    
    # PLOT A HISTOGRAM
    hist(responses,
        breaks = c(seq(-0.5, tosses+0.5, 1)),
        col = "lightblue",
        freq = FALSE,
        main = paste("Total of", tosses, "tosses"))

    # PLOT VERTICAL LINES
    abline(v = quantile(responses, prob = c(0.025, 0.975)),
          col = "orange",
          lwd = 2)

    abline(v = 1,
          col = "blue",
          lwd = 2)
}
```

We had to increase all the way to 20 or 21 for statistical significance. To decide whether it should be 20 or 21 we may need more trials. Think about how probability is defined; 10000 is a large number but probability is defined in terms of that number getting larger and larger.
